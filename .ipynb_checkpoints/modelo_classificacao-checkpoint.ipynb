### Glossário de dados:

- age: Idade (Anos)
- sex: Sexo (1 = Masc e 0 = Fem)

- trestbps: Pressão sanguínea em Repouso
- chol: colesterol em mg/dl:
- fbs: Fasting Blood Sugar (Teste diabético) > 120 mg/dl
- restecg: Eletrocardiogramas em repouso (0,1 ou 2)
- thalach: Ritmo cardíaco
- exang: Exercício físico que gerou Angina

fluoroscopia)

defect

## Construção do modelo

### Divisão treino-teste

```python
X = heart_features
y = heart_target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
model = BernoulliNB(alpha=10.0, fit_prior=True)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print('BernoulliNB')
print(f"Acurácia: {accuracy_score(y_test, y_pred)*100} %")
print(f"F1 Score: {f1_score(y_test, y_pred)*100} %")
```
